# O que é o Airbnb?
* Airbnb é um marketplace online que conecta as pessoas que necessitam de alocação com pessoas que tem um quarto, apartamento ou casa para alugar. A plataforma foi fundada em 2008 e permite que as pessoas listem as suas propriedades no site e aluguem para outras pessoas. Inquilinos podem pesquisar por propriedades baseadas em localização, preço e outros benefícios. Além disso podem pagar e agendar a sua estadia pela própria plataforma do Airbnb. O produto está presente em mais de 220 países e possui mais de 7 milhões de propriedades listadas.

* Baseado no conjunto de dados fornecido, o intuito desse projeto consiste maximizar a rentabilidade da empresa e/ou a taxa de ocupação das casas e apartamentos. Para isso usaremos uma metodologia baseada em:
1. Explorar 
2. Analisar
3. Orientar a tomada de ação

## 2.1 Análise Univariada

Na análise univariada, analisamos as distribuições de cada variável individualmente

```python
# Separando as variáveis contínuas das variáveis dicotômicas
var_num = list(df.select_dtypes(include='float64').columns)
var_cat = list(df.select_dtypes(include='object').columns)
```

```python
for col in var_cat:
    counts = df[col].value_counts(normalize=True)
    
    plt.figure(figsize=(10,8))

    counts.plot(kind='barh')
    plt.title(col, fontsize=15)
    plt.xlabel('Porcentagem',fontsize=15)
    plt.ylabel('Categorias', fontsize=15)
    plt.gca().invert_yaxis()
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.show()
```

```python
# Analisando as variáveis dicotômicas
for col in var_cat:
    counts = df[col].value_counts(normalize=True)
    counts.plot(kind='bar')
    plt.title(col)
    plt.xlabel('Categorias')
    plt.ylabel('Porcentagem')
    plt.show()
```

```python
# Analisando as variáveis numéricas
for column in var_num:
    plt.hist(df[column], bins=20, range=(0, 100), density=True, color='g', alpha=0.5)
    plt.title(column, fontsize=15)
    plt.xlabel('Valores', fontsize=15)
    plt.ylabel('Frequência', fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.show()
```

Analisando as distribuições, percebemos que as variaveis __accomodates__, __bedrooms__, __beds__ e __minimum_nights__ podem ser interpretadas como variáveis categóricas.

```python
# Criando uma lista com as novas colunas categóricas
cols_cat = var_cat + ['accommodates', 'bedrooms', 'beds', 'minimum_nights']
cols_cat
```

```python
# Criando uma lista com as novas variáveis numéricas
cols_num = list(filter(lambda x: x not in ['accommodates', 'bedrooms', 'beds', 'minimum_nights'], var_num))
cols_num
```

## 2.3 Análise Multivariada

```python
# Criação do subset de dados
df_ = df[['neighbourhood_cleansed','instant_bookable']]
df_['instant_bookable'] = df['instant_bookable'] == 't'

# Agrupamento dos dados e transferindo para tabela
df_ = df_.groupby(['neighbourhood_cleansed']).mean().reset_index().sort_values(by='instant_bookable', ascending=False)
df_
```
